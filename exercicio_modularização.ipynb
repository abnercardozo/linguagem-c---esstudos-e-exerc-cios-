```python
//CODIGO CORRETO
#include <stdio.h>

// protótipos
long long fatorial(int num);
long long arranjo(int n, int p);
long long combinacao(int n, int p);

// fatorial: n! = 1 * 2 * 3 * ... * n
long long fatorial(int num) {
    int i;
    long long fat = 1;

    for (i = 1; i <= num; i++) {
        fat = fat * i;
    }
    return fat;
}

// arranjo: A(n,p) = n! / (n-p)!
long long arranjo(int n, int p) {
    return fatorial(n) / fatorial(n - p);
}

// combinacao: C(n,p) = A(n,p) / p!
long long combinacao(int n, int p) {
    return arranjo(n, p) / fatorial(p);
}

int main() {
    int n, p;

    printf("Digite n e p (negativos para sair).\n");

    while (1) {
        printf("n = "); scanf("%d", &n);
        printf("p = "); scanf("%d", &p);

        // condição de saída
        if (n < 0 || p < 0) {
            printf("Encerrando.\n");
            break;
        }

        // validação das regras
        if (p > n) {
            printf("ERRO: n deve ser >= p!\n\n");
            continue;
        }

        printf("Arranjo(%d,%d)    = %lld\n", n, p, arranjo(n, p));
        printf("Combinacao(%d,%d) = %lld\n\n", n, p, combinacao(n, p));
    }

    return 0;
}
```

```python
//CODIGO CORRIGIDO
#include <stdio.h>

// protótipo
int miniCalculadora(int num1, int num2, char operador);

// função
int miniCalculadora(int num1, int num2, char operador) {
    switch (operador) {
        case '+':
            return num1 + num2;
        case '-':
            return num1 - num2;
        case '*':
            return num1 * num2;
        case '/':
            if (num2 == 0) {
                printf("ERRO: divisao por zero!\n");
                return 0;
            }
            return num1 / num2;
        case '%':
            if (num2 == 0) {
                printf("ERRO: divisao por zero!\n");
                return 0;
            }
            return num1 % num2;
        default:
            printf("ERRO: operador invalido!\n");
            return 0;
    }
}

int main() {
    int i, num1, num2, resultado;
    char operador;

    for (i = 0; i < 100; i++) {
        printf("Operacao %d\n", i + 1);

        printf("Numero 1: ");
        scanf("%d", &num1);

        printf("Numero 2: ");
        scanf("%d", &num2);

        printf("Operador (+, -, *, /, %%): ");
        scanf(" %c", &operador);

        resultado = miniCalculadora(num1, num2, operador);
        printf("Resultado: %d\n\n", resultado);
    }

    return 0;
}
```

```python
//CODIGO CORRIGIDO
#include <stdio.h>

// PROTOTIPO
int classificar(char simbolo);

// FUNCAO
int classificar(char simbolo) {

    // vogais minusculas — usamos '' pois se trata de um caractere
    // usamos || pois o simbolo pode ser UMA coisa OU outra
    if (simbolo == 'a' || simbolo == 'e' || simbolo == 'i' ||
        simbolo == 'o' || simbolo == 'u') {
        return 1; // VOGAL MINUSCULA

    // vogais maiusculas
    } else if (simbolo == 'A' || simbolo == 'E' || simbolo == 'I' ||
               simbolo == 'O' || simbolo == 'U') {
        return 2; // VOGAL MAIUSCULA

    // consoantes minusculas — letra minuscula que nao e' vogal
    // >= 'a' && <= 'z' forma o intervalo de letras minusculas
    } else if (simbolo >= 'a' && simbolo <= 'z') {
        return 3; // CONSOANTE MINUSCULA

    // consoantes maiusculas — letra maiuscula que nao e' vogal
    } else if (simbolo >= 'A' && simbolo <= 'Z') {
        return 4; // CONSOANTE MAIUSCULA

    // qualquer outro caractere
    } else {
        return 5; // SIMBOLO ESPECIAL
    }
}

int main() {
    int i, classe;
    char simbolo;

    // 100 SIMBOLOS
    for (i = 0; i < 100; i++) {

        // leitura do simbolo
        printf("Insira o simbolo: ");
        scanf(" %c", &simbolo);

        // CHAMAR FUNCAO
        classe = classificar(simbolo);

        // MOSTRAR RESULTADO
        printf("A classe do simbolo e: %d\n\n", classe);
    }

    return 0;
}
```

em porcentagem, retorne o preço reajustado e o reajuste calculado em reais. Após isso, use a função no
programa principal e exiba o preço reajustado e o reajuste de 100 mercadorias informadas pelo usuário.

```python
//MEU CODIGO
#include <stdio.h>
#include <stdlib.h>

//PROTOTIPO
float reajusteCalculado(float preco, float porcentagem);

//FUNCAO
float reajusteCalculado(float preco, float porcentagem){
        return (preco * porcentagem)/100;
}

int main()
{   int i;
    float preco, porcentagem, precofinal;


    //100 MERCADORIAS
    for(i=0;i<100;i++){
        printf("Insira o preço da mercadoria: \n");
        scanf("%f", &preco);

        printf("Insira o reajuste percentual da mercadoria:\n");
        scanf("%f", &porcentagem);

        //CHAMAR FUNCAO
        precofinal = reajusteCalculado(preco, porcentagem);

        printf("O preço final é: \n", &precofinal);
    }
    return 0;
}
```

```python
//CODIGO CORRIGIDO
#include <stdio.h>

// PROTOTIPO — void pois nao retorna nada pelo return
// os dois ultimos parametros sao PONTEIROS (*)
void reajusteCalculado(float preco, float porcentagem,
                       float *reajuste, float *precoFinal);

// FUNCAO — 2 caso: usa ponteiros para retornar dois valores
void reajusteCalculado(float preco, float porcentagem,
                       float *reajuste, float *precoFinal) {

    // * acessa o conteudo do endereco apontado
    // ou seja, escreve diretamente na variavel da main!
    *reajuste   = (preco * porcentagem) / 100;
    *precoFinal = preco + (*reajuste);
}

int main() {
    int i;
    float preco, porcentagem;

    // variaveis que receberao os resultados via ponteiro
    float reajuste, precoFinal;

    // 100 MERCADORIAS
    for (i = 0; i < 100; i++) {
        printf("Mercadoria %d\n", i + 1);

        printf("Insira o preco da mercadoria: ");
        scanf("%f", &preco);

        printf("Insira o reajuste percentual: ");
        scanf("%f", &porcentagem);

        // validacao
        if (preco <= 0 || porcentagem < 0) {
            printf("ERRO: valores invalidos!\n\n");
            i--;
            continue;
        }

        // CHAMAR FUNCAO — ENDERECO das variaveis com &
        // a funcao vai escrever os resultados diretamente nelas!
        reajusteCalculado(preco, porcentagem, &reajuste, &precoFinal);

        // MOSTRAR RESULTADO
        printf("Reajuste: R$ %.2f\n", reajuste);
        printf("Preco final: R$ %.2f\n\n", precoFinal);
    }

    return 0;
}
```

```python
//CODIGO CORRIGIDO
#include <stdio.h>

// PROTOTIPO — void pois retorna os valores pelos ponteiros
void converterTempo(int totalSegundos, int *h, int *m, int *s);

// FUNCAO
void converterTempo(int totalSegundos, int *h, int *m, int *s) {
    int resto;

    // divide por 3600 para obter as horas inteiras
    *h = totalSegundos / 3600;

    // resto apos tirar as horas
    resto = totalSegundos % 3600;

    // divide o resto por 60 para obter os minutos
    *m = resto / 60;

    // resto apos tirar os minutos — nao esquecer que % e' o resto e / e' a divisao!
    *s = resto % 60;
}

int main() {
    int i;
    int horas, minutos, segundos, totalSegundos;

    // USAR FUNCAO 100X
    for (i = 0; i < 100; i++) {
        printf("Entrada %d\n", i + 1);

        printf("Insira os segundos: ");
        // lemos em totalSegundos, nunca em i — i e' o contador do loop!
        scanf("%d", &totalSegundos);

        // validacao
        if (totalSegundos < 0) {
            printf("ERRO: segundos nao pode ser negativo!\n\n");
            i--;
            continue;
        }

        // CHAMANDO A FUNCAO — passamos totalSegundos como valor
        // e o ENDERECO das variaveis com & para os ponteiros escreverem nelas
        converterTempo(totalSegundos, &horas, &minutos, &segundos);

        // RESULTADO
        printf("%d Segundos = %d Hora(s) + %d Minuto(s) + %d Segundo(s)\n\n",
               totalSegundos, horas, minutos, segundos);
    }

    return 0;
}
```
