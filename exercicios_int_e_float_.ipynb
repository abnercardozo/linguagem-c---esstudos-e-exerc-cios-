```python
//CODIGO CORRIGIDO
#include <stdio.h>
#include <stdlib.h>
#define TAMANHO 15000

int main()
{
    int i, j;
    int matricula[TAMANHO];
    float tempo[TAMANHO];
    float auxFloat;
    int auxInt;

    // ETAPA 1 — CADASTRO
    for(i = 0; i < TAMANHO; i++)
    {
        printf("\n--- Atleta %d ---\n", i + 1);

        printf("Matricula: ");
        scanf("%d", &matricula[i]);

        do
        {
            printf("Tempo (min): ");
            scanf("%f", &tempo[i]);
            if(tempo[i] <= 0)
                printf("Tempo invalido! Digite valor maior que zero.\n");
        } while(tempo[i] <= 0);
    }

    // ETAPA 2 — BUBBLE SORT (ordenar pelo menor tempo)
    for(i = 0; i < TAMANHO - 1; i++)
    {
        for(j = 0; j < TAMANHO - 1 - i; j++)
        {
            if(tempo[j] > tempo[j+1])
            {
                // Troca o TEMPO
                auxFloat    = tempo[j];
                tempo[j]    = tempo[j+1];
                tempo[j+1]  = auxFloat;

                // Troca a MATRICULA junto!
                auxInt          = matricula[j];
                matricula[j]    = matricula[j+1];
                matricula[j+1]  = auxInt;
            }
        }
    }

    // ETAPA 3 — EXIBIR OS 10 MELHORES
    printf("\n=== RELATORIO 24: TOP 10 ATLETAS ===\n");
    for(i = 0; i < 10; i++)
    {
        printf("%d. Matricula: %d | Tempo: %.1f min\n",
                i + 1, matricula[i], tempo[i]);
    }

    return 0;
}
```

```python
//CODIGO CORRIGIDO
#include <stdio.h>
#include <stdlib.h>
#define TAMANHO 50

int main()
{
    int i;
    int novoCodigo;
    int pesquisa;
    int encontrou = 0;
    int codigo[TAMANHO];
    float preco[TAMANHO];
    float quantidade[TAMANHO];

    // ETAPA 1 — CADASTRO
    for(i = 0; i < TAMANHO; i++)
    {
        printf("\n--- Produto %d ---\n", i + 1);

        printf("Codigo: ");
        scanf("%d", &codigo[i]);

        do
        {
            printf("Preco (R$): ");
            scanf("%f", &preco[i]);
            if(preco[i] < 0.25)
                printf("Preco invalido! Deve ser >= R$ 0.25\n");
        } while(preco[i] < 0.25);

        do
        {
            printf("Quantidade: ");
            scanf("%f", &quantidade[i]);
            if(quantidade[i] <= 0)
                printf("Quantidade invalida! Deve ser > 0\n");
        } while(quantidade[i] <= 0);
    }

    // ETAPA 2 — PESQUISA SEQUENCIAL
    printf("\nDigite o codigo a ser alterado: ");
    scanf("%d", &pesquisa);

    for(i = 0; i < TAMANHO; i++)
    {
        if(codigo[i] == pesquisa)
        {
            printf("\n=== RELATORIO 40 ===\n");
            printf("Produto encontrado no indice: %d\n", i);
            printf("Codigo atual:    %d\n", codigo[i]);
            printf("Preco:           R$ %.2f\n", preco[i]);
            printf("Quantidade:      %.1f\n", quantidade[i]);


            printf("\nDigite o novo codigo: ");
            scanf("%d", &novoCodigo);
            codigo[i] = novoCodigo;

            printf("Codigo atualizado com sucesso!\n");
            encontrou = 1;
            break;
        }
    }

    // ETAPA 3 — SE NÃO ENCONTROU
    if(encontrou == 0)
        printf("\nCodigo %d nao encontrado!\n", pesquisa);

    return 0;
}
```
