````python
//codigo corrigido
#include <stdio.h>
#include <stdlib.h>
#include <string.h>

#define TAM 100

int main()
{
    char palavra[TAM];

    char temp[5];

    char resultado[TAM];

    // INICIALIZA STRING VAZIA
    strcpy(resultado, "");

    while(1)
    {
        printf("Digite uma palavra ou FIM para sair:\n");

        scanf("%s", palavra);

        // PARADA
        if(strcmp(palavra, "FIM") == 0)
        {
            break;
        }

        // MAIS DE 4 LETRAS
        if(strlen(palavra) > 4)
        {
            strncpy(temp, palavra, 4);

            temp[4] = '\0';

            strcat(resultado, temp);
        }
        else
        {
            strcat(resultado, palavra);
        }
    }

    // RESULTADO
    printf("\nResultado: %s\n", resultado);

    return 0;
}
````
